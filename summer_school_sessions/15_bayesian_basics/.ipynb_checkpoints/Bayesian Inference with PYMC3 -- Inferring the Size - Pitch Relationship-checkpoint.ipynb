## 1: Bayesian inference for the bias parameter of a Binomial random variable

In statistical terms, the consistent versus inconciststen responses we have been collecting from participants is known as a binomial random variable, and some standard tehcnique exist to infer the bias parameter $p$ for this variable.

In the previous excresises, we have been estimating this $p$ parameter (probability of a consistent response) by simpy taking the mean number of consistent responses. This is in fact known to be the maximum likelihood solution to the inferential problem.

However, there are lots of circumstances in which we may wish to place a prior over $p$ and performs Bayesian inference for the posterior distribution of $p$ given our experimental data and our prior. This notebook shows how to do that.

### 1.1: The Beta Binomial Model
A standard model for estimating the bias parameter of a biomial distribution is known as the Beta-Binomial model. This assumes that the prior distribution for $p$ is the family of Beta distributions. There are good statistical reasons to choose this prior, but I won;t go into those here. We'll see how to fit a Bayesian model under a Beta prior and a Uniform prior. 

```python
import pandas as pd

# grab the xprimental data
data = pd.read_csv('CleanData.csv')

# grab the pitch-size comparisons from the data
ps = data[(data.Comparison == 'Pitch-Size')]
```

We'll need to slightly reshape our data to plug it into PYMC3. For each participant, we need to know two things:

1.   How many trials did they perform (N) ?
2.   How many consistent responses did they give (k)?

PYMC3 requires that we collect this information into vectors: a vector of Ns where each entry gives the number trials a participant undertook. And a vectors of $k$s such that each entry gives the number of correct repsonses the participant undertook (in the same order as the vetor of $N$s!). 

```python
# Group the data by subject
# calling .count() counts the number of valid rows in the datafram for each subject (which gives us $N$)
# calling.values return a list of the values in this column as a numpy array
ns = ps.groupby('Subject').count()['Response'].values


# because the "Response" column is full of zeros (inconsistent) and ones (concsistent),...
# ...calling.sum() tells us how many consistent trials each participant produced (k)  
ks = ps.groupby('Subject').sum()['Response'].values
```

```python
import pymc3 as pm

with pm.Model() as model:
    
    # create a variable "p" which follows a Beta prior
    # The parameters 1,1 correspond to the parameters of a Beta distribution, and can be changed
    # see https://en.wikipedia.org/wiki/Beta_distribution for info on the Beta distirbution
    # using 1,1 makes the Beta distribution uniform on [0,1]
    p = pm.Beta('p', 1,1)
    
    # Create a Binomial variable y whcih represents the observed data
    # Provide the Ns as a parameter, and the ks as the observed "success" counts
    y = pm.Binomial('y', n = ns, p = p, observed = ks)
    
    # take 2000 samples from the posterior distirbution for p
    results = pm.sample(2000)
    
```

PYMC3 also provides some handy plotting functions for looking at the posterior...

```python
%matplotlib inline
pm.traceplot(results)
```

So we can see that with a uniform prior the posterior mean for p concentrates around .57. Try changing the parameters of the Beta prior to encode more knowledge: for example, try setting a Beta(10,10) prior, which strongly favours values of $p$ near 0.5, and see how the posterior changes. The next example shows the same model but with a uniform prior. 

```python
import pymc3 as pm

# Uniform prior model
with pm.Model() as uniform_model:
    
    # Now p follows a uniform prior
    p = pm.Uniform('p', lower = 0, upper = 1)

    y = pm.Binomial('y', n = ns, p = p, observed = ks)
    
    # take 2000 samples from the posterior distirbution for p
    uniform_results = pm.sample(2000)
```

```python
%matplotlib inline
pm.traceplot(uniform_results)
```

As well as visualising the data, the "results" object has a method to access posterior samples for any of the variables:

```python
import seaborn as sns

# grab all the posterior samples for p
p_samples = uniform_results.get_values(varname='p')

# plot the histogram of these samples
sns.distplot(p_samples)
```

The next analysis is really quite advanced: a model in which we infer a seperate $p$ parameter for every individual, but assume that these $p$ parameters are sampled from a shared distribution. In this case, we wish to learn the parameters of the shared higher order distribution. This is a hierarchical Bayesian model (in frequentist stats, this kind of model is often called "multi-level").

```python
# hierarchical model

# see tutorial from pymc3: http://docs.pymc.io/notebooks/GLM-hierarchical.html

# This utility from sklearn allows us to give a vector of unique participant identifiers
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
subject_id = le.fit_transform(ps.groupby('Subject').mean().index.values)

# how many individual $p$ parameters are we assuming ( = how many participants)
nsubjects = ps.Subject.nunique()

with pm.Model() as h_model:
    
    # parameters of the shared beta prior which we wish to infer
    # we place an exponential prior on these parameters, since the params of a beta distribution are positive real
    a = pm.Exponential('a', .01)
    b = pm.Exponential('b', .01)
    
    p = pm.Beta('p', a, b, shape = nsubjects)
    
    y = pm.Binomial('y', n = ns, p = p[subject_id], observed = ks)
    
    results_h = pm.sample(2000)
```

Let's look at these results using one of PYMC3's other plotting function, 'foresplot':

```python
pm.forestplot(results_h, varnames=['a', 'b'])
```

So the posterior mode for alpha and beta are roughly .78 and .58 respectively. Let's plot a Beta distribution with these parameters and look at the shape.

```python
from scipy.stats import beta
import numpy as np

import seaborn as sns

pdf = beta(.78, .58,).pdf(np.linspace(0,1, 1000))

sns.plt.plot(np.linspace(0,1, 1000), pdf)
```

Great, so this suggests that when we infer a population level distribution of $p$ values, we can see that we should expect most participants to have a strong bias for consistent relationships between size and pitch ($p \approx 1$) and a smaller number of subjects to demonstrate the opposite preference: a preference for inconsistent mappings ($p \approx 0$).
